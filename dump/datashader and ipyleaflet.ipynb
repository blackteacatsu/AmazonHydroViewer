```python
import xarray as xr
import geopandas as gpd
import cartopy.crs as ccrs
import json
import urllib.request
from ipyleaflet import Map, ImageOverlay, GeoJSON, GeoData, projections
from pyproj import Transformer
import datashader as dsr
import datashader.transfer_functions as tf
```

```python
ds = xr.open_dataset('/Users/kris/amazonforcast/data/forecast/output/LIS_HIST_Forecast_June_02_to_05_mean.nc')
# Extract the bounds
lat = ds["lat"].values
lon = ds["lon"].values
values = ds['Rainf_tavg'].isel(time = 1)
values
```

```python
show(values)
```

```python
# Step 2: Define a Datashader canvas
cvs = dsr.Canvas(
    plot_width=1000, plot_height=1000,  # Resolution of the output
    x_range=(lon.min(), lon.max()),  # Longitude range in Web Mercator
    y_range=(lat.min(), lat.max())   # Latitude range in Web Mercator
)

# Step 3: Rasterize the data variable
agg = cvs.points(values, x=lon, y=lat)

# Step 4: Convert to a Datashader image
img = tf.shade(agg, how="linear")

img
```
